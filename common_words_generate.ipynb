```python
def word_dict_generate(text):
    word_dict = {}
    words = []
    for line in text:
        line = line.strip().lower()
        line = re.sub(r","," ",line)
        line = re.sub(r";"," ",line)
        line = re.sub(r"\'"," ",line)
        line = re.sub(r":"," ",line)
        line = re.sub(r"!"," ",line)
        line = re.sub(r"\."," ",line)
        line = re.sub(r"\/"," ",line)
        line = re.sub(r"\?"," ", line)
        line = re.sub(r"[\\d]"," ",line)
        line = re.sub(r"\'\'"," ",line)
        line = re.sub(r"\("," ",line)
        line = re.sub(r"\)"," ",line)
        line = re.sub(r"\<"," ",line)
        line = re.sub(r"\>"," ",line)
        line = nltk.word_tokenize(line)
        for w in line:
            if w not in stopwords.words("english") and len(w)>3:
              if w not in word_dict:
                    word_dict[w]=1
              else:
                    word_dict[w]+=1
              words.append(w)
    return word_dict,words
```

```python
def text_clean(text):
    line = text.strip()
    line = re.sub(r";","",line)
    line = re.sub(r"\'","",line)
    line = re.sub(r":","",line)
    line = re.sub(r"!","",line)
    line = re.sub(r"\.","",line)
    line = re.sub(r"\/","",line)
    line = re.sub(r"\?","", line)
    line = re.sub(r"\'\'","",line)
    line = re.sub(r"\(","",line)
    line = re.sub(r"\)","",line)
    line = re.sub(r"\<","",line)
    line = re.sub(r"\>","",line)
    return line
```

```python
def computeTF(wordDict, words):
    tfDict = {}
    wordsCount = len(words)
    for word, count in wordDict.items():
        tfDict[word] = count / float(wordsCount)
    return tfDict
```

```python
def computeIDF(text):
    N = len(text)
    idfDict = dict.fromkeys(text.keys(), 0)
    for word, val in text.items():
        if val > 0:
            idfDict[word] += 1
    for word, val in idfDict.items():
        if val!=0:
            idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
def computeTFIDF(tfDict, idfs):
    tfidf = {}
    for word, val in tfDict.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
def common_words_generate(tcm_ids):
    common_words = []
    d = {}
    if len(tcm_ids) >= 2:
        a_1 = articles[articles['article_id'] == tcm_ids[0]]['text']
        d_1, w_1 = word_dict_generate(a_1)
        a_2 = articles[articles['article_id'] == tcm_ids[1]]['text']
        d_2, w_2 = word_dict_generate(a_2)
        d = {**d_1,**d_2}
        common_words = list(set(w_1).intersection(set(w_2)))
        if len(tcm_ids) > 2:
            for t in range(2,len(tcm_ids)):
                a_t = articles[articles['article_id'] == tcm_ids[t]]['text']
                d_t, w_t = word_dict_generate(a_t)
                d = {**d,**d_t}
                common_words = list(set(common_words).intersection(set(w_t)))

    top_common_words = {}
    if len(common_words):
      tf = computeTF(d,common_words)
      idf = computeIDF(d)
      tfidf = computeTFIDF(tf,idf)
      res = {}
      for w in common_words:
        res[w] = tfidf[w]

      res = sorted(res.items(), key=lambda x: x[1], reverse=True)

      for i in range(len(res)):
        if i < 10:
          top_common_words[res[i][0]] = res[i][1]

    return top_common_words, common_words
```

```python
def common_words_p_t(tcm_ids,s):
    r = []
    a_1 = articles[articles['article_id'] == tcm_ids[0]][s].to_string(index = False).strip('[').strip(']').strip().split(',')
    a_1 = [text_clean(i) for i in a_1 if i!='NaN']
    a_2 = articles[articles['article_id'] == tcm_ids[1]][s].to_string(index = False).strip('[').strip(']').strip().split(',')
    a_2 = [text_clean(i) for i in a_2 if i!='NaN']
    r = list(set(a_1).intersection(set(a_2)))
    if len(tcm_ids) > 2:
        for t in range(2,len(tcm_ids)):
            a_t = articles[articles['article_id'] == tcm_ids[t]][s].to_string(index = False).strip('[').strip(']').strip().split(',')
            a_t = [i.replace('\'','') for i in a_t]
            a_t = [text_clean(i) for i in a_t if i!='NaN']
            r = list(set(r).intersection(set(a_t)))
    return r
```
